```python
from pathlib import Path
import numpy as np
import pandas as pd

# ===== 1. Paths and basic loading =====
DATASET = Path("/space_mounts/atai-hs25/dataset")

print("DATASET exists:", DATASET.exists())
print("SUBDIRS:", list(DATASET.iterdir()))

# 1.1 Load entity IDs
entity_ids_raw = []
with (DATASET / "embeddings" / "entity_ids.del").open("r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split("\t")
        if len(parts) == 2:
            idx, uri = parts
            entity_ids_raw.append(uri)

uri2idx = {uri: i for i, uri in enumerate(entity_ids_raw)}

print("\n# Entities")
print("num_entities:", len(entity_ids_raw))
print("first 5 entity URIs:", entity_ids_raw[:5])

# 1.2 Load entity embeddings
E = np.load(DATASET / "embeddings" / "entity_embeds.npy")
print("\n# Entity embeddings")
print("E.shape:", E.shape, "dtype:", E.dtype)

# Precompute norms for cosine similarity
E_norms = np.linalg.norm(E, axis=1) + 1e-8


# ===== 2. Load movie_plots.csv (correct path) =====
plots_path = DATASET / "additional" / "movie_plots.csv"
plots = pd.read_csv(plots_path)

print("\n# movie_plots.csv")
print("path:", plots_path)
print("shape:", plots.shape)
print("columns:", list(plots.columns))
print(plots.head())

# rename column for clarity: index -> uri
if "index" in plots.columns:
    plots = plots.rename(columns={"index": "uri"})

print("\nAfter rename:")
print(plots.head())


# ===== 3. Search for our test movies / keywords in plot text =====
keywords = [
    "Lion King",
    "Pocahontas",
    "Beauty and the Beast",
    "Nightmare on Elm Street",
    "Friday the 13th",
    "Halloween",
]

print("\n# Keyword search in plots")
for kw in keywords:
    mask = plots["plot"].str.contains(kw, case=False, na=False)
    sub = plots[mask].head(5)
    print(f"\n== Keyword: {kw!r} ==")
    print("matches:", mask.sum())
    if not sub.empty:
        print(sub[["uri", "plot"]])
    else:
        print("(no rows found)")


# ===== 4. Pick a 'beast'-like movie as a concrete example =====
print("\n# Searching for a 'gentle-hearted beast' style movie in plots")
mask_beast = plots["plot"].str.contains("gentle-hearted beast", case=False, na=False)
beast_df = plots[mask_beast]

if beast_df.empty:
    print("No 'gentle-hearted beast' plot found. Showing a generic 'beast' example instead.")
    mask_beast = plots["plot"].str.contains("beast", case=False, na=False)
    beast_df = plots[mask_beast].head(1)

if beast_df.empty:
    print("Still no 'beast' example found, cannot continue with this part.")
    beast_uri = None
else:
    print("\nCandidate 'beast' movie rows:")
    print(beast_df[["uri", "plot"]].head(3))

    beast_uri = beast_df["uri"].iloc[0]
    print("\nChosen beast_uri:", beast_uri)

# Build uri -> plot dict for movies (for display)
uri2plot = dict(zip(plots["uri"], plots["plot"]))


# ===== 5. Nearest neighbors in teacher embedding for this beast_uri =====
def top_k_neighbors(uri, k=10):
    if uri not in uri2idx:
        print(f"URI not found in entity_ids: {uri}")
        return []
    idx = uri2idx[uri]
    v = E[idx]
    v_norm = np.linalg.norm(v) + 1e-8

    # cosine similarity
    sims = (E @ v) / (E_norms * v_norm)

    # get top k+1 (including itself)
    topk = np.argsort(-sims)[:k+1]

    results = []
    for j in topk:
        if j == idx:
            continue
        nb_uri = entity_ids_raw[j]
        sim = float(sims[j])
        plot = uri2plot.get(nb_uri, "<no plot / not in movie_plots>")
        if isinstance(plot, str) and len(plot) > 120:
            plot_short = plot[:120] + "..."
        else:
            plot_short = plot
        results.append((j, nb_uri, sim, plot_short))
        if len(results) >= k:
            break
    return results

if beast_uri is not None:
    print("\n# Top-10 neighbors for beast_uri in teacher embedding")
    neighbors = top_k_neighbors(beast_uri, k=10)
    for j, nb_uri, sim, plot_short in neighbors:
        print(f"idx={j:6d} | sim={sim: .4f} | uri={nb_uri}")
        print(f"  plot: {plot_short}")
else:
    print("\nNo beast_uri available, skipping neighbor exploration.")


# ===== 6. Look at KG triples that involve this beast_uri =====
def inspect_triples_for_uri(uri, max_subject=20, max_object=20):
    """Print triples where this uri appears as subject or object in graph.tsv"""
    if uri is None:
        print("No uri given.")
        return

    graph_path = DATASET / "graph.tsv"
    subj_triples = []
    obj_triples = []

    # The URIs in graph.tsv are wrapped like <http://www.wikidata.org/entity/Q...>
    needle = f"<{uri}>"

    with graph_path.open("r", encoding="utf-8") as f:
        for line in f:
            parts = line.rstrip("\n").split("\t")
            if len(parts) != 3:
                continue
            s, p, o = parts
            if s == needle:
                subj_triples.append((s, p, o))
                if len(subj_triples) >= max_subject and len(obj_triples) >= max_object:
                    break
            elif o == needle:
                obj_triples.append((s, p, o))
                if len(subj_triples) >= max_subject and len(obj_triples) >= max_object:
                    break

    print(f"\n# Triples with {needle} as subject (up to {max_subject})")
    if subj_triples:
        for s, p, o in subj_triples:
            print(s, p, o)
    else:
        print("(none)")

    print(f"\n# Triples with {needle} as object (up to {max_object})")
    if obj_triples:
        for s, p, o in obj_triples:
            print(s, p, o)
    else:
        print("(none)")

if beast_uri is not None:
    inspect_triples_for_uri(beast_uri, max_subject=20, max_object=20)
else:
    print("\nSkipping triple inspection because no beast_uri was selected.")
```

```python
from pathlib import Path
import pandas as pd

# --- Paths ---

DATASET = Path("/space_mounts/atai-hs25/dataset")
EMB_DIR = DATASET / "embeddings"
ADDITIONAL_DIR = DATASET / "additional"
GRAPH_TSV = DATASET / "graph.tsv"

OUT_DIR = Path("/files/Final Project Submission/metadata")
OUT_DIR.mkdir(parents=True, exist_ok=True)

print("DATASET:", DATASET)
print("OUT_DIR:", OUT_DIR)

# --- 1) Backup entity_ids.del as a clean TSV ---

entity_ids_path = EMB_DIR / "entity_ids.del"
entity_rows = []

with entity_ids_path.open("r", encoding="utf-8") as f:
    for line in f:
        line = line.rstrip("\n")
        if not line:
            continue
        parts = line.split("\t", 1)
        if len(parts) != 2:
            continue
        idx_str, uri = parts
        try:
            idx = int(idx_str)
        except ValueError:
            continue
        entity_rows.append((idx, uri))

entity_rows.sort(key=lambda x: x[0])

entity_ids_out = OUT_DIR / "entity_ids_ordered.tsv"
with entity_ids_out.open("w", encoding="utf-8") as f:
    f.write("index\turi\n")
    for idx, uri in entity_rows:
        f.write(f"{idx}\t{uri}\n")

print(f"Saved entity_ids_ordered.tsv with {len(entity_rows)} rows")


# --- 2) Backup relation_ids.del as a clean TSV ---

relation_ids_path = EMB_DIR / "relation_ids.del"
relation_rows = []

with relation_ids_path.open("r", encoding="utf-8") as f:
    for line in f:
        line = line.rstrip("\n")
        if not line:
            continue
        parts = line.split("\t", 1)
        if len(parts) != 2:
            continue
        idx_str, uri = parts
        try:
            idx = int(idx_str)
        except ValueError:
            continue
        relation_rows.append((idx, uri))

relation_rows.sort(key=lambda x: x[0])

relation_ids_out = OUT_DIR / "relation_ids_ordered.tsv"
with relation_ids_out.open("w", encoding="utf-8") as f:
    f.write("index\trelation_uri\n")
    for idx, uri in relation_rows:
        f.write(f"{idx}\t{uri}\n")

print(f"Saved relation_ids_ordered.tsv with {len(relation_rows)} rows")


# --- 3) Backup movie_plots.csv as TSV (uri + plot) ---

movie_plots_path = ADDITIONAL_DIR / "movie_plots.csv"
plots_df = pd.read_csv(movie_plots_path)
# normalize column names
plots_df = plots_df.rename(columns={"index": "uri"})
plots_out = OUT_DIR / "movie_plots.tsv"
plots_df.to_csv(plots_out, sep="\t", index=False)
print(f"Saved movie_plots.tsv with shape {plots_df.shape}")


# --- 4) Detect film entities from graph.tsv: P31 == Q11424 (instance of film) ---

film_uri_set = set()
p_instance_of = "<http://www.wikidata.org/prop/direct/P31>"
film_class = "<http://www.wikidata.org/entity/Q11424>"

with GRAPH_TSV.open("r", encoding="utf-8") as f:
    for line in f:
        line = line.rstrip("\n")
        if not line:
            continue
        try:
            s, p, o = line.split("\t")
        except ValueError:
            continue
        if p == p_instance_of and o == film_class:
            # strip angle brackets if present
            uri = s.strip("<>")
            film_uri_set.add(uri)

print(f"Detected {len(film_uri_set)} film URIs from graph.tsv")

plots_uris = set(plots_df["uri"])

# Build a film_entities.tsv aligned with entity_ids_ordered.tsv
film_entities_out = OUT_DIR / "film_entities.tsv"
with film_entities_out.open("w", encoding="utf-8") as f:
    f.write("index\turi\tis_film\thas_plot\n")
    for idx, uri in entity_rows:
        # teacher entity_ids may or may not have angle brackets; normalize
        clean_uri = uri.strip("<>")
        is_film = "1" if clean_uri in film_uri_set else "0"
        has_plot = "1" if clean_uri in plots_uris else "0"
        f.write(f"{idx}\t{uri}\t{is_film}\t{has_plot}\n")

print("Saved film_entities.tsv (index, uri, is_film, has_plot)")


# --- 5) Extract rdfs:label titles for film / plot entities from graph.tsv ---

label_pred = "<http://www.w3.org/2000/01/rdf-schema#label>"
title_uris_of_interest = film_uri_set | plots_uris

titles = {}  # uri -> title string

with GRAPH_TSV.open("r", encoding="utf-8") as f:
    for line in f:
        line = line.rstrip("\n")
        if not line:
            continue
        try:
            s, p, o = line.split("\t")
        except ValueError:
            continue
        if p != label_pred:
            continue
        uri = s.strip("<>")
        if uri not in title_uris_of_interest:
            continue

        # object is a literal: "Title"@en or "Title"^^type etc.
        o_str = o.strip()
        if not o_str.startswith('"'):
            continue
        # naive parse: take content between first two quotes
        parts = o_str.split('"')
        if len(parts) < 3:
            continue
        text = parts[1]
        # prefer first label we see; if you want, you can later add language filtering
        if uri not in titles:
            titles[uri] = text

print(f"Collected {len(titles)} titles with rdfs:label")

titles_out = OUT_DIR / "entity_titles.tsv"
with titles_out.open("w", encoding="utf-8") as f:
    f.write("uri\ttitle\n")
    for uri, title in titles.items():
        safe_title = title.replace("\t", " ").replace("\n", " ")
        f.write(f"{uri}\t{safe_title}\n")

print("Saved entity_titles.tsv (uri, title)")
```

```python
from pathlib import Path

import numpy as np

# --- Paths ---

DATASET = Path("/space_mounts/atai-hs25/dataset")
GRAPH_TSV = DATASET / "graph.tsv"

META_DIR = Path("/files/Final Project Submission/metadata")
OUT_TRIPLES_PATH = META_DIR / "rfc_triples_ids.tsv"

print("DATASET:", DATASET)
print("GRAPH_TSV:", GRAPH_TSV)
print("META_DIR:", META_DIR)

# --- 1) Load entity and relation ID mappings (teacher order) ---

entity_ids_file = META_DIR / "entity_ids_ordered.tsv"
relation_ids_file = META_DIR / "relation_ids_ordered.tsv"

uri2eid = {}
with entity_ids_file.open("r", encoding="utf-8") as f:
    header = next(f)  # skip header
    for line in f:
        idx_str, uri = line.rstrip("\n").split("\t")
        eid = int(idx_str)
        uri2eid[uri] = eid

print(f"Loaded entity mapping: {len(uri2eid)} entities")

reluri2rid = {}
with relation_ids_file.open("r", encoding="utf-8") as f:
    header = next(f)  # skip header
    for line in f:
        idx_str, uri = line.rstrip("\n").split("\t")
        rid = int(idx_str)
        reluri2rid[uri] = rid

print(f"Loaded relation mapping: {len(reluri2rid)} relations")


# --- 2) Convert graph.tsv to (sid, rid, oid) triples using teacher IDs ---

def strip_brackets(u: str) -> str:
    # graph.tsv subjects/objects are usually like <http://...>
    # but entity_ids_ordered.tsv URIs are stored without < >
    u = u.strip()
    if u.startswith("<") and u.endswith(">"):
        return u[1:-1]
    return u

num_lines = 0
num_triples = 0
num_skipped_literal = 0
num_skipped_missing = 0

# we will write directly to a TSV file: sid \t rid \t oid
with OUT_TRIPLES_PATH.open("w", encoding="utf-8") as out_f, GRAPH_TSV.open("r", encoding="utf-8") as in_f:
    out_f.write("sid\trid\toid\n")
    for line in in_f:
        num_lines += 1
        line = line.rstrip("\n")
        if not line:
            continue
        try:
            s, p, o = line.split("\t")
        except ValueError:
            # malformed line
            num_skipped_missing += 1
            continue

        s_uri = strip_brackets(s)
        p_uri = strip_brackets(p)

        # if object starts with a quote, it's a literal (e.g. "Title"@en), skip
        if o.strip().startswith('"'):
            num_skipped_literal += 1
            continue
        o_uri = strip_brackets(o)

        # look up IDs
        if s_uri not in uri2eid or o_uri not in uri2eid or p_uri not in reluri2rid:
            num_skipped_missing += 1
            continue

        sid = uri2eid[s_uri]
        rid = reluri2rid[p_uri]
        oid = uri2eid[o_uri]

        out_f.write(f"{sid}\t{rid}\t{oid}\n")
        num_triples += 1

print("Done converting graph.tsv")
print(f"Total lines read:       {num_lines}")
print(f"Valid triples written:  {num_triples}")
print(f"Skipped literals:       {num_skipped_literal}")
print(f"Skipped missing IDs:    {num_skipped_missing}")
print(f"Output triples file:    {OUT_TRIPLES_PATH}")
```

## Define TransE model and training utilities

```python
# CELL 2: Define TransE model and training utilities

import torch
from torch import nn

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

class TransEModel(nn.Module):
    def __init__(self, num_entities, num_relations, emb_dim=256, margin=1.0):
        super().__init__()
        self.emb_dim = emb_dim
        self.margin = margin

        self.entity_emb = nn.Embedding(num_entities, emb_dim)
        self.relation_emb = nn.Embedding(num_relations, emb_dim)

        # Simple initialization
        nn.init.uniform_(self.entity_emb.weight, a=-0.01, b=0.01)
        nn.init.uniform_(self.relation_emb.weight, a=-0.01, b=0.01)

    def forward(self, h, r, t):
        # h, r, t are LongTensor indices
        h_e = self.entity_emb(h)
        r_e = self.relation_emb(r)
        t_e = self.entity_emb(t)
        # TransE score: L2 distance of h + r - t
        return torch.norm(h_e + r_e - t_e, p=2, dim=-1)

    def loss_fn(self, pos_score, neg_score):
        # margin ranking loss: max(0, margin + pos - neg)
        return torch.relu(self.margin + pos_score - neg_score).mean()


def generate_negative_batch(h_batch, r_batch, t_batch, num_entities):
    # Randomly corrupt head or tail
    batch_size = h_batch.shape[0]
    # 0 = corrupt head, 1 = corrupt tail
    corrupt_head = torch.randint(0, 2, (batch_size,), device=h_batch.device)

    random_entities = torch.randint(0, num_entities, (batch_size,), device=h_batch.device)

    h_neg = torch.where(corrupt_head == 1, h_batch, random_entities)
    t_neg = torch.where(corrupt_head == 1, random_entities, t_batch)

    return h_neg, r_batch, t_neg
```

## Train TransE and save RFC embeddings

```python
# CELL 3: Train TransE and save RFC embeddings

from math import ceil

# Move triples to device once
h_all = sid.to(device)
r_all = rid.to(device)
t_all = oid.to(device)

num_triples = h_all.shape[0]
print("Training triples:", num_triples)

emb_dim = 256
margin = 1.0
batch_size = 8192  # adjust if you hit OOM
epochs = 10
lr = 1e-3

model = TransEModel(num_entities=num_entities,
                    num_relations=num_relations,
                    emb_dim=emb_dim,
                    margin=margin).to(device)

optimizer = torch.optim.Adam(model.parameters(), lr=lr)

steps_per_epoch = ceil(num_triples / batch_size)
print("steps_per_epoch:", steps_per_epoch)

model.train()
for epoch in range(1, epochs + 1):
    perm = torch.randperm(num_triples, device=device)
    h_shuffled = h_all[perm]
    r_shuffled = r_all[perm]
    t_shuffled = t_all[perm]

    total_loss = 0.0

    for step in range(steps_per_epoch):
        start = step * batch_size
        end = min((step + 1) * batch_size, num_triples)
        if start >= end:
            break

        h_batch = h_shuffled[start:end]
        r_batch = r_shuffled[start:end]
        t_batch = t_shuffled[start:end]

        # Positive scores
        pos_score = model(h_batch, r_batch, t_batch)

        # Negative samples
        h_neg, r_neg, t_neg = generate_negative_batch(h_batch, r_batch, t_batch, num_entities)
        neg_score = model(h_neg, r_neg, t_neg)

        loss = model.loss_fn(pos_score, neg_score)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

        # optional: print every N steps
        if (step + 1) % 100 == 0:
            avg_loss = total_loss / (step + 1)
            print(f"Epoch {epoch} Step {step+1}/{steps_per_epoch} - avg_loss: {avg_loss:.4f}", end="\r")

    avg_epoch_loss = total_loss / steps_per_epoch
    print(f"\nEpoch {epoch} finished - avg_loss: {avg_epoch_loss:.4f}")

# After training, move embeddings to CPU and save as RFC_entity_embeds.npy
RFC_EMB_DIR = Path("/files/Final Project Submission/embeddings")
RFC_EMB_DIR.mkdir(parents=True, exist_ok=True)

entity_emb_matrix = model.entity_emb.weight.detach().cpu().numpy()
print("entity_emb_matrix shape:", entity_emb_matrix.shape)

out_path = RFC_EMB_DIR / "RFC_entity_embeds.npy"
np.save(out_path, entity_emb_matrix)
print("Saved RFC embeddings to:", out_path)
```

```python
import numpy as np
import pandas as pd
from pathlib import Path

# ---------------------------
# Paths & basic loading
# ---------------------------
DATASET_ROOT = Path("/space_mounts/atai-hs25/dataset")
RFC_EMB_PATH = Path("/files/Final Project Submission/embeddings/RFC_entity_embeds.npy")
MOVIE_PLOTS_PATH = DATASET_ROOT / "additional" / "movie_plots.csv"
ENTITY_IDS_PATH = DATASET_ROOT / "embeddings" / "entity_ids.del"

print("DATASET_ROOT:", DATASET_ROOT)
print("RFC_EMB_PATH exists:", RFC_EMB_PATH.exists())
print("MOVIE_PLOTS_PATH exists:", MOVIE_PLOTS_PATH.exists())
print("ENTITY_IDS_PATH exists:", ENTITY_IDS_PATH.exists())

# --- load entity ids ---
uris = []
with open(ENTITY_IDS_PATH, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split("\t")
        if len(parts) == 2:
            uris.append(parts[1])

num_entities = len(uris)
print(f"\nLoaded {num_entities} entity URIs")

uri_to_idx = {u: i for i, u in enumerate(uris)}

# --- load RFC embedding ---
E = np.load(RFC_EMB_PATH)
print("RFC embedding shape:", E.shape)

# --- load plots for inspection (optional, just for printing) ---
plots_df = pd.read_csv(MOVIE_PLOTS_PATH)
plots_df.rename(columns={"index": "uri"}, inplace=True)
uri_to_plot = dict(zip(plots_df["uri"], plots_df["plot"]))

def qid_to_uri(qid: str) -> str:
    """Convert 'Q123' -> 'http://www.wikidata.org/entity/Q123'"""
    qid = qid.strip()
    if qid.startswith("Q"):
        return f"http://www.wikidata.org/entity/{qid}"
    return qid  # already full URI


def get_plot(uri: str, max_len: int = 140) -> str:
    """Return short plot snippet if available."""
    p = uri_to_plot.get(uri, "<no plot in movie_plots>")
    p = str(p)
    return (p[: max_len - 3] + "...") if len(p) > max_len else p


def cosine_sim(a: np.ndarray, b: np.ndarray) -> np.ndarray:
    """Compute cosine similarity between vector(s) a and all b."""
    # a: (d,) or (n, d), b: (m, d)
    if a.ndim == 1:
        a = a[None, :]
    # normalize
    a_norm = a / (np.linalg.norm(a, axis=1, keepdims=True) + 1e-9)
    b_norm = b / (np.linalg.norm(b, axis=1, keepdims=True) + 1e-9)
    return a_norm @ b_norm.T  # (n, m)


def evaluate_group(name: str, qids, top_k: int = 10):
    """
    For a semantic group (e.g. '1950s musicals'), do:
      1) map QIDs -> indices
      2) compute intra-group cosine sim stats
      3) compute centroid -> nearest neighbours in whole KG
    """
    print("\n" + "=" * 80)
    print(f"GROUP: {name}")
    print("-" * 80)

    qids = [q.strip() for q in qids if q.strip()]
    qid_uris = [qid_to_uri(q) for q in qids]

    # map to indices
    found = []
    missing = []
    for q, u in zip(qids, qid_uris):
        idx = uri_to_idx.get(u)
        if idx is None:
            missing.append((q, u))
        else:
            found.append((q, u, idx))

    print(f"Total QIDs: {len(qids)}")
    print(f"Found in entity_ids: {len(found)}")
    print(f"Missing in entity_ids: {len(missing)}")

    if missing:
        print("  Missing examples (up to 5):")
        for q, u in missing[:5]:
            print(f"    {q} -> {u} (NOT in mapping)")

    if len(found) < 2:
        print("  [SKIP] Less than 2 seeds found, cannot compute group stats.")
        return

    # --- intra-group cosine similarity ---
    idxs = np.array([t[2] for t in found], dtype=int)
    group_vecs = E[idxs]          # (n, d)
    S = cosine_sim(group_vecs, group_vecs)  # (n, n)

    # exclude self-sim when computing stats
    n = S.shape[0]
    mask = ~np.eye(n, dtype=bool)
    sims = S[mask]

    print("\nIntra-group cosine similarity stats (excluding self):")
    print(f"  mean: {sims.mean():.4f}")
    print(f"  min : {sims.min():.4f}")
    print(f"  max : {sims.max():.4f}")

    # --- centroid neighbours ---
    centroid = group_vecs.mean(axis=0)
    sims_all = cosine_sim(centroid, E)[0]  # (num_entities,)
    top_idx = np.argsort(-sims_all)[:top_k]

    # Set of URIs in group for quick membership test
    group_uri_set = {u for _, u, _ in found}

    print(f"\nTop-{top_k} neighbours of group centroid:")
    in_group_cnt = 0
    for rank, idx in enumerate(top_idx, start=1):
        uri = uris[idx]
        sim = sims_all[idx]
        # try to recover QID if it's a wikidata URI
        qid = uri.rsplit("/", 1)[-1] if "wikidata.org/entity/" in uri else "<non-WD>"
        mark = "[IN-GROUP]" if uri in group_uri_set else "          "
        if mark.strip():
            in_group_cnt += 1

        print(f"{rank:3d}. sim={sim:.4f} {mark}")
        print(f"     idx={idx} uri={uri} qid={qid}")
        print("     plot:", get_plot(uri))
    print(f"\nGroup members among top-{top_k}: {in_group_cnt}/{top_k}")


# ---------------------------
# Define semantic groups from lecture queries
# ---------------------------

# 1) 1950s musicals (from the prompt list)
musicals_1950s_qids = [
    "Q1198780", "Q2061652", "Q3497184", "Q7750889", "Q96212867", "Q189875",
    "Q747570", "Q2249349", "Q2556456", "Q584509", "Q3548988", "Q146141",
    "Q586782", "Q2249423", "Q2407907", "Q7743045", "Q1508130", "Q1210771",
    "Q1032943104726858", "Q7731955299242752", "Q270470", "Q1615122",
    "Q2208041", "Q284362", "Q383355", "Q5021556", "Q1824959",
    "Q7114058", "Q5901227107430037", "Q1228404",
]

# 2) Fellini movies
fellini_qids = [
    "Q18433", "Q18416", "Q18405", "Q18441", "Q18439", "Q18428", "Q18415",
    "Q12018", "Q3600277797414023", "Q18429", "Q18437", "Q18411",
    "Q18402", "Q18423", "Q18399",
]

# 3) Italian drama films (Q130232)
italian_drama_qids = [
    "Q2617110", "Q1637024", "Q991511", "Q535295", "Q643827",
    "Q1635122", "Q249782", "Q598994", "Q1212325",
]

# 4) Colleen Atwood filmography (partial list from prompt)
colleen_atwood_qids = [
    "Q204057", "Q18199331", "Q1783958", "Q1195568", "Q288173",
    "Q278997", "Q208592", "Q133654", "Q116905", "Q752989",
    "Q10858674", "Q1058555", "Q14847067", "Q17485699", "Q323318",
    "Q850522", "Q271690", "Q886719", "Q639481", "Q217182",
    "Q629975", "Q624603", "Q18199330", "Q4021832045956022",
    "Q19321019", "Q687627", "Q212041", "Q329448",
    "Q5814321120170702", "Q469624", "Q1355172",
]

# 5) Costume Design Oscar films
costume_oscar_qids = [
    "Q17738", "Q202211", "Q24284363", "Q1757288",
    "Q207921", "Q829695", "Q471716", "Q3521099",
]

# 6) Toru Takemitsu films
takemitsu_qids = ["Q1634355", "Q1207936", "Q3232745", "Q1659440"]

# 7) Other Japanese films (Q5287)
japanese_film_qids = [
    "Q1634715", "Q1631964", "Q1632477", "Q2342125172251179",
    "Q107656066", "Q1316093", "Q3549482", "Q1458055",
    "Q2739665", "Q2718554",
]

# 8) Meryl Streep biographical movies
meryl_bio_qids = [
    "Q19955874", "Q269810", "Q380848", "Q30203425",
    "Q1166180", "Q208266", "Q1117936", "Q1249239",
    "Q1412037",
]

# ---------------------------
# Run evaluations
# ---------------------------

evaluate_group("1950s musicals", musicals_1950s_qids, top_k=15)
evaluate_group("Fellini movies", fellini_qids, top_k=15)
evaluate_group("Italian drama films", italian_drama_qids, top_k=15)
evaluate_group("Colleen Atwood filmography", colleen_atwood_qids, top_k=15)
evaluate_group("Costume Design Oscar films", costume_oscar_qids, top_k=15)
evaluate_group("Toru Takemitsu films", takemitsu_qids, top_k=15)
evaluate_group("Japanese films (general)", japanese_film_qids, top_k=15)
evaluate_group("Meryl Streep biographical movies", meryl_bio_qids, top_k=15)
```

```python
import numpy as np
import pandas as pd
from numpy.linalg import norm
from pathlib import Path

base = Path(".")

EMB_PATH = base / "embeddings" / "RFC_entity_embeds.npy"
ENTITY_IDS_PATH = base / "metadata" / "entity_ids_ordered.tsv"
TITLES_PATH = base / "metadata" / "entity_titles.tsv"
FILM_ENTITIES_PATH = base / "metadata" / "film_entities.tsv"

# 1) 加载 embeddings
entity_embeds = np.load(EMB_PATH)
print("entity_embeds:", entity_embeds.shape)

# 2) 加载 entity_ids，并强制对齐到 embeddings 的行数
entity_ids_df = pd.read_csv(ENTITY_IDS_PATH, sep="\t", dtype=str)
print("entity_ids_df:", entity_ids_df.shape)
print(entity_ids_df.head())

extra = entity_ids_df.shape[0] - entity_embeds.shape[0]
if extra > 0:
    print(f"→ entity_ids 比 embeddings 多 {extra} 行，删掉开头这几行以对齐")
    entity_ids_df = entity_ids_df.iloc[extra:].reset_index(drop=True)

entity_ids = entity_ids_df.iloc[:, 0].astype(str).tolist()
assert len(entity_ids) == entity_embeds.shape[0], (
    f"len(entity_ids)={len(entity_ids)} vs embeds={entity_embeds.shape[0]}"
)
print("num entity_ids:", len(entity_ids))

# 3) 加载 titles，并对齐到 entity_ids
titles_df = pd.read_csv(TITLES_PATH, sep="\t", dtype=str)
print("titles_df:", titles_df.shape)
print(titles_df.head())

extra_titles = titles_df.shape[0] - len(entity_ids)
if extra_titles > 0:
    print(f"→ titles 比 entity_ids 多 {extra_titles} 行，删掉开头这几行以对齐")
    titles_df = titles_df.iloc[extra_titles:].reset_index(drop=True)

titles_df.iloc[:, 0] = titles_df.iloc[:, 0].astype(str)

# 4) 加载 film_entities
film_df = pd.read_csv(FILM_ENTITIES_PATH, sep="\t", dtype=str)
print("film_df:", film_df.shape)
print(film_df.head())

film_eids = film_df.iloc[:, 0].astype(str).tolist()
print("num film_entities (raw):", len(film_eids))

# 5) 建立 eid → index 映射，并检查覆盖率
eid2idx = {eid: i for i, eid in enumerate(entity_ids)}
covered = [eid for eid in film_eids if eid in eid2idx]
print("film_entities covered by embeddings:", len(covered))

def top_k_neighbors_for_eid(eid, k=10):
    """给一个 entity_id，返回最相似的 k 个实体（不含自己）"""
    if eid not in eid2idx:
        print("EID not in index:", eid)
        return []
    idx = eid2idx[eid]
    v = entity_embeds[idx]
    sims = entity_embeds @ v / (norm(entity_embeds, axis=1) * norm(v) + 1e-9)
    sims[idx] = -1.0  # 排除自己
    k = min(k, len(sims) - 1)
    top_idx = np.argpartition(-sims, k)[:k]
    top_idx = top_idx[np.argsort(-sims[top_idx])]
    results = []
    for j in top_idx:
        neighbor_eid = entity_ids[j]
        row = titles_df[titles_df.iloc[:, 0] == neighbor_eid]
        title = row.iloc[0, 1] if not row.empty else "(no title)"
        results.append((float(sims[j]), neighbor_eid, title))
    return results

# 6) 选一个有 embedding 的电影实体，打印邻居
if covered:
    sample_eid = covered[0]
    print("\nSample film entity:", sample_eid)
    row = titles_df[titles_df.iloc[:, 0] == sample_eid]
    sample_title = row.iloc[0, 1] if not row.empty else "(no title)"
    print("Title:", sample_title)

    neighbors = top_k_neighbors_for_eid(sample_eid, k=10)
    print("\nTop neighbors:")
    for score, eid, title in neighbors:
        print(f"{score:.3f} | {eid} | {title}")
else:
    print("⚠ 没有任何 film_entities 能在 embeddings 中找到，请检查对齐问题。")
```

```python
import numpy as np
import pandas as pd

print("entity_embeds:", entity_embeds.shape)
print("entity_ids_df (raw):", entity_ids_df.shape)
print("film_df (raw):", film_df.shape)
print("titles_df (raw):", titles_df.shape)

max_embed_idx = entity_embeds.shape[0] - 1

# --- 小工具：安全清理 index 列 ---
def clean_index_column(df, col="index"):
    df = df.copy()
    # 把非数字的都变成 NaN（例如 'index' 这种表头）
    df[col] = pd.to_numeric(df[col], errors="coerce")
    before = len(df)
    df = df.dropna(subset=[col])
    dropped = before - len(df)
    if dropped > 0:
        print(f"[INFO] dropped {dropped} non-numeric rows from '{col}'")
    df[col] = df[col].astype(int)
    return df.reset_index(drop=True)

# 1) 清理 entity_ids_df 和 film_df 的 index 列
entity_ids_df = clean_index_column(entity_ids_df, "index")
film_df = clean_index_column(film_df, "index")

# 2) 过滤掉没有 embedding 的实体（index >= len(entity_embeds)）
film_df = film_df[film_df["index"] <= max_embed_idx].copy()
film_entity_ids = film_df.loc[film_df["is_film"] == 1, "index"].astype(int).to_numpy()

print("entity_ids_df (clean):", entity_ids_df.shape)
print("film_df (clean):", film_df.shape)
print("num film_entities (covered by embeddings):", film_entity_ids.shape[0])

# 3) 清理 titles_df，只保留真正的 URI 行
titles_df_clean = titles_df[titles_df["uri"].astype(str).str.startswith("http")].copy()
titles_df_clean = titles_df_clean.reset_index(drop=True)
print("titles_df (clean):", titles_df_clean.shape)
print("examples from titles_df_clean:")
print(titles_df_clean.head())

# 4) 建立各种映射

# entity_id -> uri
idx_to_uri = dict(zip(entity_ids_df["index"].astype(int), entity_ids_df["uri"].astype(str)))

# uri -> title
title_by_uri = dict(zip(titles_df_clean["uri"].astype(str), titles_df_clean["title"].astype(str)))

# 有 title 且有 embedding 的所有实体（不限制 film）
all_titled_entity_ids = []
for eid in idx_to_uri.keys():
    if eid <= max_embed_idx:
        uri = idx_to_uri[eid]
        if uri in title_by_uri:
            all_titled_entity_ids.append(eid)
all_titled_entity_ids = sorted(all_titled_entity_ids)
print("num titled entities with embeddings (all types):", len(all_titled_entity_ids))

# 有 title 且是 film 的实体
titled_film_ids = [eid for eid in film_entity_ids if idx_to_uri.get(int(eid)) in title_by_uri]
titled_film_ids = sorted(set(titled_film_ids))
print("num titled film entities with embeddings:", len(titled_film_ids))

# 决定用哪一批作为“种子实体”
if len(titled_film_ids) > 0:
    seed_ids = titled_film_ids
    restrict_neighbors_to_films = True
    print("\n[INFO] Using titled film entities as seeds.")
elif len(all_titled_entity_ids) > 0:
    seed_ids = all_titled_entity_ids
    restrict_neighbors_to_films = False
    print("\n[INFO] No titled film entities found; using all titled entities as seeds.")
else:
    print("\n⚠️ No entities with titles and embeddings found at all.")
    print("   这说明 entity_titles.tsv 基本是空的，当前数据没法做有意义的语义探索。")
    seed_ids = []
    restrict_neighbors_to_films = False

# 5) 归一化 embedding，用余弦相似度
if len(seed_ids) > 0:
    norms = np.linalg.norm(entity_embeds, axis=1, keepdims=True)
    norms[norms == 0] = 1.0
    entity_embeds_norm = entity_embeds / norms

    def top_k_neighbors_cosine(idx, k=10, restrict_to_ids=None):
        """返回 (sim, j, uri_j, title_j) 列表"""
        v = entity_embeds_norm[idx]
        sims = entity_embeds_norm @ v  # 余弦相似度

        if restrict_to_ids is not None:
            mask = np.full(len(sims), False)
            restrict_to_ids = np.asarray(restrict_to_ids, dtype=int)
            restrict_to_ids = restrict_to_ids[restrict_to_ids <= max_embed_idx]
            mask[restrict_to_ids] = True
            sims = np.where(mask, sims, -np.inf)

        k = min(k, len(sims))
        topk = np.argpartition(-sims, range(k))[:k]
        topk = topk[np.argsort(-sims[topk])]

        results = []
        for j in topk:
            uri_j = idx_to_uri.get(int(j), "<missing uri>")
            title_j = title_by_uri.get(uri_j, "(no title)")
            results.append((float(sims[j]), int(j), uri_j, title_j))
        return results

    print("\n================ EXPLORATION ================\n")
    # 只探索前几个，避免输出爆炸
    for eid in seed_ids[:5]:
        uri = idx_to_uri.get(int(eid), "<missing uri>")
        title = title_by_uri.get(uri, "(no title)")
        print(f"=== Entity {eid} ===")
        print("URI  :", uri)
        print("Title:", title)

        if restrict_neighbors_to_films:
            neighbors = top_k_neighbors_cosine(eid, k=10, restrict_to_ids=film_entity_ids)
            print("Top neighbors among film entities (cosine similarity):")
        else:
            neighbors = top_k_neighbors_cosine(eid, k=10, restrict_to_ids=None)
            print("Top neighbors among ALL entities (cosine similarity):")

        for sim, j, uri_j, title_j in neighbors:
            print(f"{sim: .3f} | idx={j:6d} | {uri_j} | {title_j}")
        print()
```

```python
import numpy as np
import pandas as pd
import textwrap
import random

# ========= 0. 路径配置（如有不同，只改这里） =========
EMB_PATH           = "embeddings/RFC_entity_embeds.npy"   # 如果你叫 entity_embeddings.npy 就改成那个
ENTITY_IDS_PATH    = "metadata/entity_ids_ordered.tsv"
FILM_ENTITIES_PATH = "metadata/film_entities.tsv"
MOVIE_PLOTS_PATH   = "metadata/movie_plots.tsv"

# ========= 1. 加载 embeddings =========
if "entity_embeds" not in globals():
    entity_embeds = np.load(EMB_PATH)
print("entity_embeds:", entity_embeds.shape)

# ========= 2. 读取并清洗 metadata =========
def load_entity_ids(path):
    df = pd.read_csv(path, sep="\t", dtype=str, low_memory=False)
    # 把 header 混进去的那一行干掉
    bad = (df["uri"] == "uri") | (df["index"] == "index")
    df = df[~bad].copy()

    df["index"] = pd.to_numeric(df["index"], errors="coerce")
    df = df.dropna(subset=["index"])
    df["index"] = df["index"].astype(int)

    df = df.sort_values("index").reset_index(drop=True)

    # 如果比 embedding 行数多，就裁掉
    if len(df) > len(entity_embeds):
        df = df.iloc[:len(entity_embeds)].copy()

    print("entity_ids_df:", df.shape)
    return df


def load_film_entities(path):
    df = pd.read_csv(path, sep="\t", dtype=str, low_memory=False)
    bad = (df["uri"] == "uri") | (df["index"] == "index")
    df = df[~bad].copy()

    df["index"] = pd.to_numeric(df["index"], errors="coerce")
    df = df.dropna(subset=["index"])
    df["index"] = df["index"].astype(int)
    df = df.sort_values("index").reset_index(drop=True)

    for col in ("is_film", "has_plot"):
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0).astype(int)

    n_films = int(df["is_film"].sum()) if "is_film" in df.columns else -1
    print("film_df:", df.shape, "| #films:", n_films)
    return df


def load_movie_plots(path):
    # 不信 header，直接当「无表头」读进来，然后自己命名列
    df_raw = pd.read_csv(path, sep="\t", header=None, dtype=str, low_memory=False)
    print("movie_plots_raw shape:", df_raw.shape)

    n_cols = df_raw.shape[1]
    if n_cols >= 3:
        # 取前 3 列：uri, title, plot
        df = df_raw.iloc[:, :3].copy()
        df.columns = ["uri", "title", "plot"]
    elif n_cols == 2:
        # 只有 uri + plot，没有 title
        df = df_raw.iloc[:, :2].copy()
        df.columns = ["uri", "plot"]
        df["title"] = ""           # 塞一个空 title
        df = df[["uri", "title", "plot"]]
    else:
        raise ValueError(f"movie_plots.tsv 只有 {n_cols} 列，至少需要 uri + plot 两列")

    # 去掉 header 那一行（如果第一行是 'uri' 那种）
    df = df[df["uri"].astype(str).str.startswith("http")].copy()

    print("movie_plots_df:", df.shape)
    return df


entity_ids_df   = load_entity_ids(ENTITY_IDS_PATH)
film_df         = load_film_entities(FILM_ENTITIES_PATH)
movie_plots_df  = load_movie_plots(MOVIE_PLOTS_PATH)

# ========= 3. 建立映射 & 只保留「有 plot 的电影」 =========
idx_to_uri = dict(zip(entity_ids_df["index"], entity_ids_df["uri"]))
uri_to_idx = {u: i for i, u in idx_to_uri.items()}

# 只保留 is_film=1 的行，并且 index < len(embeddings)
film_only = film_df[film_df["is_film"] == 1].copy()
film_only = film_only[film_only["index"] < len(entity_embeds)].copy()

# 合并剧情简介（按 uri）
film_merged = film_only.merge(movie_plots_df, on="uri", how="left")

# 只保留有 plot 的
film_merged = film_merged[film_merged["plot"].notna()].copy()

# 映射：电影 idx -> 文本描述
film_indices = film_merged["index"].astype(int).tolist()
uri_to_plot  = dict(zip(film_merged["uri"], film_merged["plot"].astype(str)))

print("num film entities:", film_only.shape[0])
print("num film entities with description+embedding:", len(film_indices))

# ========= 4. 预先单位化 embeddings，用 cosine similarity =========
emb = entity_embeds
norm = np.linalg.norm(emb, axis=1, keepdims=True)
norm[norm == 0] = 1.0
emb_norm = emb / norm

film_indices_arr = np.array(film_indices, dtype=int)

# ========= 5. 主函数：给一部电影，找相似电影 =========
def recommend_similar_films(uri, k=10):
    """给一个电影 URI，打印它的简介和 top-k 相似电影（按 embedding 相似度）。"""
    if uri not in uri_to_idx:
        print("URI not in entity_ids:", uri)
        return

    idx = uri_to_idx[uri]
    if idx not in film_indices:
        print("This URI is not marked as film (or has no plot):", uri)
        return

    v = emb_norm[idx]

    # 只在「电影子集」里算相似度
    sims = emb_norm[film_indices_arr] @ v

    # 排序，排除自己
    order = np.argsort(-sims)
    top = []
    for pos in order:
        j = int(film_indices_arr[pos])
        if j == idx:
            continue  # 跳过自己
        top.append((j, sims[pos]))
        if len(top) >= k:
            break

    print("\n=== QUERY FILM ===")
    print("URI :", uri)
    print("idx :", idx)
    print("Desc:", textwrap.shorten(uri_to_plot.get(uri, "(no plot)"),
                                   width=220, placeholder="..."))

    print("\nTop neighbors among films:")
    for j, s in top:
        u = idx_to_uri.get(j, "(missing uri)")
        desc = uri_to_plot.get(u, "(no plot)")
        desc_short = textwrap.shorten(desc, width=220, placeholder="...")
        print(f"{s:6.3f} | idx={j:6d} | {desc_short}")


# ========= 6. 小 demo：随机看看几部电影的近邻 =========
random.seed(42)

print("\n\n===== DEMO: sample 3 films and show neighbors =====")
sample_uris = random.sample(list(film_merged["uri"]), k=3)
for u in sample_uris:
    recommend_similar_films(u, k=8)
    print("\n" + "="*80 + "\n")
```

```python
import numpy as np
import pandas as pd
import textwrap

# ========= 1. 选出 movie_plots 里的文本列（plot/description） =========
# 假设 movie_plots_df 已经在内存中了
cols = list(movie_plots_df.columns)

# 第一列大概率是 uri，剩下的列里找文本列
non_uri_cols = [c for c in cols if c != "uri"]

plot_col = None
# 优先按常见命名猜一遍
for cand in ["plot", "description", "desc", "text", "summary"]:
    if cand in non_uri_cols:
        plot_col = cand
        break

# 如果没猜中，就退而求其次：拿非 uri 的第一列当文本
if plot_col is None:
    if len(non_uri_cols) == 0:
        raise ValueError("movie_plots_df 里除了 'uri' 之外没有别的列，无法找到剧情文本列。")
    plot_col = non_uri_cols[0]

print(f"[INFO] using text column in movie_plots_df: {plot_col!r}")

# 去掉没有 uri 或没有文本的行
movie_plots_clean = movie_plots_df.dropna(subset=["uri", plot_col]).copy()

# ========= 2. 准备 entity 映射 =========
# entity_ids_df: 必须有 'index' 和 'uri' 列，并且 index 是 int，和 embedding 行号对应
if entity_ids_df["index"].dtype != np.int64 and entity_ids_df["index"].dtype != np.int32:
    entity_ids_df["index"] = pd.to_numeric(entity_ids_df["index"], errors="coerce")
    entity_ids_df = entity_ids_df.dropna(subset=["index"])
    entity_ids_df["index"] = entity_ids_df["index"].astype(int)

# 只保留合法 index
entity_ids_df = entity_ids_df[(entity_ids_df["index"] >= 0) &
                              (entity_ids_df["index"] < len(entity_embeds))].copy()

idx_to_uri = dict(zip(entity_ids_df["index"], entity_ids_df["uri"]))
uri_to_idx = {u: i for i, u in idx_to_uri.items()}

# ========= 3. 只保留「有文本 + 有 embedding 的电影」 =========
# film_df 里有 'uri' 和 'is_film'
film_with_text = film_df.merge(
    movie_plots_clean[["uri", plot_col]],
    on="uri",
    how="inner"
)

# 只保留 is_film == 1 的
if "is_film" in film_with_text.columns:
    film_with_text = film_with_text[film_with_text["is_film"] == 1].copy()

# 映射到 embedding index
film_with_text["idx"] = film_with_text["uri"].map(uri_to_idx)
film_with_text = film_with_text.dropna(subset=["idx"]).copy()
film_with_text["idx"] = film_with_text["idx"].astype(int)

# 只保留 idx 在 embedding 范围内的
film_with_text = film_with_text[
    (film_with_text["idx"] >= 0) &
    (film_with_text["idx"] < len(entity_embeds))
].reset_index(drop=True)

film_indices = film_with_text["idx"].values
print("num films with text + embedding:", len(film_indices))

# ========= 4. 单位化 embedding，用 cosine similarity =========
if "emb_norm" not in globals():
    emb = entity_embeds
    norm = np.linalg.norm(emb, axis=1, keepdims=True)
    norm[norm == 0] = 1.0
    emb_norm = emb / norm
    print("[INFO] computed emb_norm")


def _short_text(text, width=160):
    """把剧情文本缩短到一行，方便打印。"""
    if not isinstance(text, str):
        text = str(text)
    return textwrap.shorten(" ".join(text.split()), width=width, placeholder="...")


# ========= 5. 按关键词搜索电影 =========
def search_films(keyword, n=5):
    """
    在剧情文本里搜索 keyword，返回前 n 部电影，并打印 URI / idx / 剧情摘要。
    """
    mask = film_with_text[plot_col].astype(str).str.contains(keyword, case=False, na=False)
    hits = film_with_text[mask].head(n)

    if hits.empty:
        print(f"[search_films] No films found for keyword: {keyword!r}")
        return hits  # 方便你后面自己用

    print(f"[search_films] keyword={keyword!r}, found={mask.sum()}, showing top {len(hits)}")
    for _, row in hits.iterrows():
        uri = row["uri"]
        idx = row["idx"]
        desc = _short_text(row[plot_col])
        print("-" * 80)
        print("URI :", uri)
        print("idx :", idx)
        print("Text:", desc)
    print("-" * 80)
    return hits


# ========= 6. 给定 URI，看相似电影 =========
def show_neighbors_for_film_uri(uri, k=10):
    """
    给定某个电影的 URI，在「有文本的电影」集合里找 top-k 相似电影。
    打印：相似度 / idx / URI / 剧情摘要。
    """
    if uri not in uri_to_idx:
        print("[show_neighbors_for_film_uri] URI not in entity_ids mapping:", uri)
        return

    idx = uri_to_idx[uri]
    v = emb_norm[idx]

    # 只在有文本的电影集合里算相似度
    sims = emb_norm[film_indices] @ v

    # 找 top k+1（可能包含自己），再去掉自己
    if len(sims) <= 1:
        print("[show_neighbors_for_film_uri] Not enough films with text.")
        return

    k_eff = min(k, len(sims) - 1)
    top_idx_in_film = np.argpartition(-sims, k_eff)[: k_eff + 1]

    # 去掉自己
    top_idx_in_film = [i for i in top_idx_in_film if film_indices[i] != idx][:k_eff]
    ordered = sorted(top_idx_in_film, key=lambda i: -sims[i])

    # 打印 query 本身
    src_rows = film_with_text[film_with_text["uri"] == uri]
    if not src_rows.empty:
        src_text = _short_text(src_rows.iloc[0][plot_col])
    else:
        src_text = "(no text available)"

    print("\n=== QUERY FILM ===")
    print("URI :", uri)
    print("idx :", idx)
    print("Desc:", src_text)
    print("\nTop neighbors among films:\n")

    # 打印邻居
    for i in ordered:
        j = int(film_indices[i])
        n_uri = idx_to_uri.get(j, "(missing uri)")
        n_row = film_with_text[film_with_text["uri"] == n_uri]
        if not n_row.empty:
            n_text = _short_text(n_row.iloc[0][plot_col])
        else:
            n_text = "(no text)"

        print(f"{sims[i]:6.3f} | idx={j:6d} | {n_text}")


print("\n[READY] You can now use:")
print("  search_films('India', n=5)")
print("  show_neighbors_for_film_uri('http://www.wikidata.org/entity/XXXXXX', k=10)")
```
